```python
import os
import re
```

```python
os.listdir()
```

```python
with open("row.txt","r",encoding="utf-8") as file:
    a=file.read()
    
```

```python
b=re.findall(r"[аА][бБ0]",a)
print (b)
```

```python
b=re.findall(r"[аА]б{2,3}",a)
print (b)
```

```python
b=re.findall(r"[а-яА-Я]+_+[а-яА-Я]+\S",a)
print (b)
```

```python
b=re.findall(r"[А-Я]+[а-я]+\S",a)
print (b)
```

```python
b=re.findall(r".*а.*б$",a)
print (b)
```

```python
b=re.sub(r"([,.]| )",lambda x: x.group(0).replace(',',':').replace('.',':').replace(' ',':'),a)
print (b)
```

```python
b=re.findall(r"([А-Я]|[A-Z])",a)
print (b)
```

```python
#ex7
b = re.sub(r" (\w)", lambda x: x.group(1).upper(), a).replace(' ', '')
print(b)
```
